# Calculer la similarité entre deux mots

```python
import numpy as np

# Extract the vector representations for the words "Morocco" and "Marrakech"
morocco_vector = doc[doc.text.lower().split().index('morocco')].vector
marrakech_vector = doc[doc.text.lower().split().index('marrakech')].vector

# Calculate cosine similarity
similarity = morocco_vector.dot(marrakech_vector) / (np.linalg.norm(morocco_vector) * np.linalg.norm(marrakech_vector))

print("Similarity between 'Morocco' and 'Marrakech':", similarity)
```

# Extraire les mots contextuels (les plus similaires) pour un mot central donné

```python

# Get the similarity scores between the central word "Morocco" and other words in the text
central_word = 'morocco'
similarity_scores = {word.text: word.similarity(doc[doc.text.lower().split().index(central_word)]) for word in doc}

similar_words = sorted(similarity_scores, key=similarity_scores.get, reverse=True)

print(f"Top 5 similar words to '{central_word}':", similar_words[1:6])  # Exclude the word itself
```
